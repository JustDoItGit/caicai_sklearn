```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

## 自己创建数据集

```python
X, y = make_blobs(n_samples=500, n_features=2, centers=4, random_state=1)
```

```python
X.shape
```

```python
y.shape
```

```python
fig, ax1 = plt.subplots(1)
ax1.scatter(X[:, 0], X[:, 1],  # .scatter散点图
            marker='o',  # 点的形状
            s=8  # 点的大小
            )
plt.show()
```

```python
# 如果我们想要看见这个点的分布，怎么办？
color = ['red', 'pink', 'orange', 'gray']

fig, ax1 = plt.subplots(1)
for i in range(4):
    i_index = y == i
    ax1.scatter(X[i_index, 0], X[i_index, 1],  # .scatter散点图
                marker='o',  # 点的形状
                s=8,  # 点的大小
                c=color[i]
                )
plt.show()
```

## 对数据进行聚类

```python
from sklearn.cluster import KMeans
n_clusters = 3
```

```python
cluster = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
```

```python
# 重要属性Labels_，查看聚好的类别，每个样本所对应的类
y_pred = cluster.labels_
y_pred
```

```python
# KMeans因为并不需要建立模型或者预测结果，因此我们只需要fit就能够得到聚类结果了
# KMeans也有接口predict和fit_predict，表示学习数据X并对X的类进行预测
# 但所得到的结果和我们不调用predict，直接fit之后调用属性labels一模一伴
pre = cluster.fit_predict(X)
pre
```

```python
(pre == y_pred).sum()  # 全都是True
```

```python
# 我们什么时候需要predict呢？当数据量太大的时候！
# 其实我们不必使用所有的数据来寻找质心，少量的数据就可以帮助我们确定质心了
# 当我们数据量非常大的时候，我们可以使用部分数据来帮助我们确认质心
# 剩下的数据的聚类结果，使用predict来调用
cluster_smallsub = KMeans(n_clusters=n_clusters, random_state=0).fit(X[:200])
```

```python
y_pred_ = cluster_smallsub.predict(X)
```

```python
y_pred_
```

```python
(y_pred_ == y_pred).sum()
# 数据量非常大的时候，效果会好
# 但从运行得出这样的结果，肯定与直接fit全部数据会不一致。有时候，当我们不要求那么精确，或者我们的数据量实在太大，那我们可以使用这种方法，使用接口predict
# 如果数据量还行，不是特别大，直接使用fit之后调用属性.labels_提出来
```

```python
(y_pred_ == y).sum()
```

### 重要属性cLuster_centers_，查看质心

```python
# 重要属性cLuster_centers_，查看质心
centroid = cluster.cluster_centers_
centroid
```

### 重要属性inertia_，查看总距离平方和

```python
# 重要属性inertia_，查看总距离平方和
inertia = cluster.inertia_
inertia
```

```python
color = ["red", "pink", "orange", "gray"]

fig, ax1 = plt.subplots(1)

for i in range(n_clusters):
    ax1.scatter(X[y_pred == i, 0], X[y_pred == i, 1],
                marker='o',  # 点的形状
                s=8,  # 点的大小
                c=color[i]
                )

ax1.scatter(centroid[:, 0], centroid[:, 1], marker="x", s=15, c="black")
plt.show()
```

```python
# 如果我们把猜测的羡数换成4，Inertia会怎么样？
n_clusters = 4
cluster_ = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
inertia_ = cluster_.inertia_
inertia_
```

```python
n_clusters = 5
cluster_ = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
inertia_ = cluster_.inertia_
inertia_
```

```python
n_clusters = 6
cluster_ = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
inertia_ = cluster_.inertia_
inertia_
```

```python
cluster_.score(X)
```

```python
%%time
n_clusters = 500
cluster_ = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
inertia_ = cluster_.inertia_
inertia_
```

```python
cluster_.score(X)
```

## 聚类算法的模型评估指标

### 当真实标签未知的时候：轮廓系数

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import silhouette_samples
```

```python
X.shape
```

```python
y_pred
```

```python
silhouette_score(X, y_pred)
```

```python
n_clusters = 4
cluster_ = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
silhouette_score(X, cluster_.labels_)
```

```python
n_clusters = 5
cluster_ = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
silhouette_score(X, cluster_.labels_)
```

```python
n_clusters = 6
cluster_ = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
silhouette_score(X, cluster_.labels_)
```

```python
silhouette_samples(X, y_pred)
silhouette_samples(X, y_pred).shape  # (500,)
silhouette_samples(X, y_pred).mean()
```

### 当真实标签未知的时候：Calinski-Harabaz Index

```python
from sklearn.metrics import calinski_harabasz_score
```

```python
calinski_harabasz_score(X, y_pred)
```
